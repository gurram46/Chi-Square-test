# Association between Device Type and Customer Satisfaction


**Background:**

Mizzare Corporation has collected data on customer satisfaction levels for two types of smart home devices: Smart Thermostats and Smart Lights. They want to determine if there's a significant association between the type of device purchased and the customer's satisfaction level.


**Data Provided:**

The data is summarized in a contingency table showing the counts of customers in each satisfaction level for both types of devices:
Screenshot (23).png




**Objective:**

To use the Chi-Square test for independence to determine if there's a significant association between the type of smart home device purchased (Smart Thermostats vs. Smart Lights) and the customer satisfaction level.


Assignment Tasks:
1. State the Hypotheses:
2. Compute the Chi-Square Statistic:
3. Determine the Critical Value:
Using the significance level (alpha) of 0.05 and the degrees of freedom (which is the number of categories minus 1)
4. Make a Decision:
Compare the Chi-Square statistic with the critical value to decide whether to reject the null hypothesis.


# The Hypotheses:
The goal is to clarify the relationship between device type and satisfaction level, so we start by defining hypotheses:

Null Hypothesis (H₀): There is no association between the device type (Smart Thermostat or Smart Light) and customer satisfaction levels.

Alternative Hypothesis (H₁): There is an association between the device type and customer satisfaction levels.

# Chisquare
First, you'll need to calculate the expected frequencies for each cell in the contingency table

```python
import pandas as pd
import numpy as np
from scipy import stats as st
```

we are importing pandas , numpy & stats from scipy for calucating chisquare

```python
data = {'Satisfaction': ['Very Satisfied', 'Satisfied', 'Neutral', 'Unsatisfied', 'Very Unsatisfied'],
        'Smart Thermostat': [50, 80, 60, 30, 20],
        'Smart Light': [70, 100, 90, 50, 50]}
```

Now we are creating a data frame using given data




we are creating DataFrame to data

```python
df=pd.DataFrame(data)
```

Seting Satifaction as index to create proper contingency table format

```python
contingency_table = df.set_index('Satisfaction')
```

So we are created a contingecy table and here it is

```python
contingency_table
```

Now we are calucalting the chi-square Test

So we are importing chi-square from scipy for calucalting the chi-square

```python
from scipy.stats import chi2_contingency
chi2_stat,p,dof,expected=chi2_contingency(contingency_table)
```

chi2_stat is chi-square,

p= p-value,

dof(Degrees of Freedom)=(number of row-1)*(number of columns-1)

expected frequecny=(Row Total*Columns)/Grand Total



```python
print(chi2_stat,p,dof,expected)
```

# Determine the Critical Value:

Given

Using the significance level (alpha) of 0.05 and the degrees of freedom (which is the number of categories minus 1)


```python
alpha=0.05
```

To calculate the critical value we have to use stats.chi2.ppf where 1-alpha and df

```python
critical_value=st.chi2.ppf(1-alpha, dof)
```

```python
critical_value
```

critical value is = 9.487729036781154

# Make a Decision:




Compare the Chi-Square statistic with the critical value to decide whether to reject the null hypothesis.



Since the Chi-Square statistic (5.638) is less than the critical value (9.4877), we fail to reject the null hypothesis.

So,

Ch-Square < Critical value

5.638 < 9.4877

So we Fail to reject Null Hypothesis

**Null Hypothesis (H₀)**: There is no association between the device type (Smart Thermostat or Smart Light) and customer satisfaction levels.

```python
if chi2_stat > critical_value:
  print("Reject null hypothesis")
else:
  print("Fail to reject null hypothesis")
```

# Null Hypothesis (H₀): There is no association between the device type (Smart Thermostat or Smart Light) and customer satisfaction levels.
